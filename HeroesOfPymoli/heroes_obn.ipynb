```python
#import dependencies
import pandas as pd
```

```python
#load csv file
input_file = "purchase_data.csv"
pymoli_data = pd.read_csv(input_file)
pymoli_data.head()
```

```python
#player count
players = pymoli_data["SN"].unique()
player_count = len(players)

player_count_summary = pd.DataFrame({"Total Player Count":[player_count]})
player_count_summary.head()
```

```python
#purchasing analysis (total)
items = pymoli_data["Item Name"].unique()
item_count = len(items)

avg_purch_price = pymoli_data["Price"].mean()

total_purchases = pymoli_data["Purchase ID"].count()

total_revenue = pymoli_data["Price"].sum()

purchasing_summary_total = pd.DataFrame({"Number of Items":[item_count],"Average Purchase Price":[avg_purch_price],"Total Purchases":[total_purchases],"Total Revenue":[total_revenue]})

purchasing_summary_total["Average Purchase Price"] = purchasing_summary_total["Average Purchase Price"].map("${:,.2f}".format)
purchasing_summary_total["Total Revenue"] = purchasing_summary_total["Total Revenue"].map("${:,.2f}".format)

purchasing_summary_total
```

```python
#gender demographics
grouped_gender = pymoli_data.groupby(["Gender"])
player_count_gender = grouped_gender["SN"].nunique()

gender_summary = pd.DataFrame({"Player Count":player_count_gender})
gender_summary["Percentage"] = gender_summary["Player Count"] / player_count * 100

gender_summary = gender_summary.sort_values("Percentage",ascending=False)

gender_summary["Percentage"] = gender_summary["Percentage"].map("{:.2f}%".format)

gender_summary
```

```python
#purchasing analysis by gender
purchase_count_gender = grouped_gender["Purchase ID"].count()
purchase_total_gender = grouped_gender["Price"].sum()

gender_purchase_summary = pd.DataFrame({"Purchase Count":purchase_count_gender,"Purchase Total":purchase_total_gender,"Player Count":player_count_gender})

gender_purchase_summary["Average Purchase Total"] = gender_purchase_summary["Purchase Total"] / gender_purchase_summary["Purchase Count"]
gender_purchase_summary["Average Purchase per Person"] = gender_purchase_summary["Purchase Total"] / gender_purchase_summary["Player Count"]

del gender_purchase_summary["Player Count"]

gender_purchase_summary["Purchase Total"] = gender_purchase_summary["Purchase Total"].map("${:,.2f}".format)
gender_purchase_summary["Average Purchase Total"] = gender_purchase_summary["Average Purchase Total"].map("${:,.2f}".format)
gender_purchase_summary["Average Purchase per Person"] = gender_purchase_summary["Average Purchase per Person"].map("${:,.2f}".format)

gender_purchase_summary
```

```python
#age demographics
age_categories = [0,9,14,19,24,29,34,39,999]
age_names = ["<10","10-14","15-19","20-24","25-29","30-34","35-39","40+"]

pymoli_data["Age Range"] = pd.cut(pymoli_data["Age"],age_categories,labels=age_names)

grouped_age = pymoli_data.groupby("Age Range")
player_count_age = grouped_age["SN"].nunique()

age_summary = pd.DataFrame({"Player Count":player_count_age})

age_summary["Percentage of Players"] = age_summary["Player Count"] / player_count * 100
age_summary["Percentage of Players"] = age_summary["Percentage of Players"].map("{:.2f}%".format)

age_summary
```

```python
#purchasing analysis by age
#purchasing analysis by gender
purchase_count_age = grouped_age["Purchase ID"].count()
purchase_total_age = grouped_age["Price"].sum()

age_purchase_summary = pd.DataFrame({"Purchase Count":purchase_count_age,"Purchase Total":purchase_total_age,"Player Count":player_count_age})

age_purchase_summary["Average Purchase Total"] = age_purchase_summary["Purchase Total"] / age_purchase_summary["Purchase Count"]
age_purchase_summary["Average Purchase per Person"] = age_purchase_summary["Purchase Total"] / age_purchase_summary["Player Count"]

del age_purchase_summary["Player Count"]

age_purchase_summary["Purchase Total"] = age_purchase_summary["Purchase Total"].map("${:,.2f}".format)
age_purchase_summary["Average Purchase Total"] = age_purchase_summary["Average Purchase Total"].map("${:,.2f}".format)
age_purchase_summary["Average Purchase per Person"] = age_purchase_summary["Average Purchase per Person"].map("${:,.2f}".format)

age_purchase_summary
```

```python
#top spenders
grouped_player = pymoli_data.groupby(["SN"])

player_purchase_count = grouped_player["Purchase ID"].count()
player_purchase_value = grouped_player["Price"].sum()

player_spending_summary = pd.DataFrame({"Purchase Count": player_purchase_count,"Total Purchase Value":player_purchase_value})

player_spending_summary["Average Purchase Value"] = player_spending_summary["Total Purchase Value"] / player_spending_summary["Purchase Count"]

player_spending_summary = player_spending_summary.sort_values("Total Purchase Value",ascending=False)

player_spending_summary["Total Purchase Value"] = player_spending_summary["Total Purchase Value"].map("${:.2f}".format)
player_spending_summary["Average Purchase Value"] = player_spending_summary["Average Purchase Value"].map("${:.2f}".format)

player_spending_summary.iloc[0:5,:]
```

```python
#most popular items
grouped_item = pymoli_data.groupby(["Item ID","Item Name"])

item_purchase_count = grouped_item["Purchase ID"].count()
item_purchase_value = grouped_item["Price"].sum()
item_purchase_price = grouped_item["Price"].first()

item_summary = pd.DataFrame({"Purchase Count":item_purchase_count,
                             "Price":item_purchase_price,
                            "Total Purchase Value":item_purchase_value})

item_summary = item_summary.sort_values("Total Purchase Value",ascending=False)

item_summary["Price"] = item_summary["Price"].map("${:.2f}".format)
item_summary["Total Purchase Value"] = item_summary["Total Purchase Value"].map("${:.2f}".format)

item_summary.head()
```
